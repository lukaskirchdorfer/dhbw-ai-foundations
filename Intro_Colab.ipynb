## 1.3 Visualizations

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 100 + np.random.randn(200)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 100, where=(ys > 100), facecolor='b', alpha=0.7)

plt.title("Example Visualization")
plt.show()
```
